# 판다(pandas)

판다는 r의 dataframe과 같이 테이블 형태의 데이터를 분석하는 기본툴이다. 개인적의 r dataframe보다 더 많은 기능을 담고 있어서 편리하다. 

## Pandas series& Data Frame 
판다의 기본 단위는 series와 dataframe으로 구성된다.
* series 는 index로 구성된 행이다. 
* dataframe은 series의 합으로써, index(가로행)과 column(세로열)로 구성되어있다. 또한 shape,index,colunm와 같은 인스턴스를 같는다. series가  dataframe으로 합해지면, series는 열병합을 한다. 

```python
import pandas as pd
names=['Bob','Jessica','Mary','John','Mel']
births=[968,155,77,578,973]
BabyDataSet=list(zip(names,births))

df=pd.DataFrame(data=BabyDataSet, columns=['Names','Births'])
print(df);print(df.shape);print(df.index);print(df.columns);print(df.axes)
```

데이터 프레임은 dictionary형태로 데이터의 직접 입력이 가능하며 concat을 통해서 merge가 가능하다. 

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                   index=[8, 9, 10, 11])

 
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])

frames=[df1,df2,df3]
result1=pd.concat(frames)
result2=pd.concat(frames,keys=['x','y','z'])
result3=pd.concat([df2,df4],axis=1)
result3
```
